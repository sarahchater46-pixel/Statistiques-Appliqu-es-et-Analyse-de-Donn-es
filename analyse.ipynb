```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('phase2.csv')
```

```python
df
```

```python
moyenne = df['Spending_USD'].mean()
mediane = df['Spending_USD'].median()
variance = df['Spending_USD'].var()
ecart_type = df['Spending_USD'].std()

print(f"Moyenne : {moyenne}")
print(f"Médiane : {mediane}")
print(f"Variance : {variance}")
print(f"Écart-type : {ecart_type}")
```
