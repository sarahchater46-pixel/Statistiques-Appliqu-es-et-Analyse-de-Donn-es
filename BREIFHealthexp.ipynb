```python
import numpy as np
observes = np.array([87, 76, 62, 19]) 
```

```python
# 2. Les proportions du patron
prop_patron = np.array([0.4, 0.3, 0.2, 0.1])
```

```python
total_observe = observes.sum()
theoriques = total_observe * prop_patron
```

```python
from scipy.stats import chisquare
stat, p_value = chisquare(f_obs=observes, f_exp=theoriques)

print(f"P-value : {p_value:.4f}")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# 1. Créer la table de contingence (Fréquences)
table = pd.crosstab(df['day'], df['size'])
print(table)
```

```python
# 2. Exécuter le test du Khi-deux
chi2, p_val, dof, expected = chi2_contingency(table)

print(f"Statistique Khi-deux : {chi2:.4f}")
print(f"P-value : {p_val:.4f}")
```

```python


# Filtrer uniquement pour les voitures américaines (USA)
df_usa = df[df['origin'] == 'usa']

# Calculer la moyenne du poids par année
weight_trend = df_usa.groupby('model_year')['weight'].mean().reset_index()

# Création du graphique
plt.figure(figsize=(10, 6))
sns.lineplot(data=weight_trend, x="model_year", y="weight", marker="o", color="red", linewidth=2.5)

# Personnalisation
plt.title("Évolution du poids moyen des voitures américaines (1970-1982)", fontsize=14)
plt.xlabel("Année du modèle")
plt.ylabel("Poids moyen (lbs)")
plt.grid(True, linestyle='--', alpha=0.7)

```

```python
# Création du graphique
plt.figure(figsize=(12, 6))

# On observe la répartition du nombre de cylindres par année
sns.countplot(data=df, x="model_year", hue="cylinders", palette="magma")

# Personnalisation
plt.title("Répartition du nombre de cylindres par année du modèle (1970-1982)", fontsize=14)
plt.xlabel("Année du modèle (19xx)")
plt.ylabel("Nombre de véhicules")
plt.legend(title="Nb Cylindres", loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
```
